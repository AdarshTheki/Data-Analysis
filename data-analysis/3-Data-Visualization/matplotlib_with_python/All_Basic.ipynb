```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
```

```python
plt.figure(figsize = (4,3))
plt.plot ([1,2,3,4],[1,4,9,16])
plt.title("First plot")
plt.xlabel("x - label")
plt.ylabel("y- label")
plt.xticks( fontsize = "15", rotation = 45)
plt.show()
```

```python
plt.figure(figsize = (4,3))
x = np.arange(1,5)
y = x**3
plt.plot ([1,2,3,4],[1,4,9,16],"go", x, y, 'r^') # r^ = red symbole
plt.title("2nd Plot")
plt.show()
```

```python
plt.figure(figsize = (4,3))
plt.subplot(1,2,1)
plt.plot ([1,2,3,4],[1,4,9,16],"go")
plt.title("1st title")

plt.subplot(1,2,2)
plt.plot ([1,2,3,4],[1,4,9,16],"go")
plt.title("2nd title")

plt.suptitle("My sub-title")
plt.show()
```

```python
plt.figure(figsize = (4,3))
plt.subplot(3,1,1)
plt.plot ([1,2,3,4],[1,4,9,16],"go")
plt.title("1st title")

plt.subplot(3,1,2)
plt.plot ([1,2,3,4],[1,4,9,16],"r^")
plt.title("2nd title")

plt.subplot(3,1,3)
plt.plot ([1,2,3,4],[1,4,9,16],"r^")
plt.title("3rd title")

plt.suptitle("My sub-title")
plt.show()
```

```python
x = np.arange(1,5)
y = x**3
fig, ax = plt.subplots(nrows=2, ncols=2, figsize = (6,6)) 
ax[0,1].plot([1,2,3,4], [1,4,9,16], "go")
ax[1,0].plot(x,y,"r^")
ax[0,1].set_title("Square of 1")
ax[1,0].set_title("Cubes of 2")
plt.show()
```

```python
# Bar Charts
plt.figure(figsize = (4,3))
Div = ["Div-A","Div-B","Div-C","Div-C","Div-D","Div-E",]
Div_av_marks = [70,82,73,65,68,30]
plt.bar(Div, Div_av_marks, color = "green")
plt.title("Bar Graph", fontsize = 15, weight='bold')
plt.xlabel("Divisions", weight='bold')
plt.ylabel("Marks", weight='bold')
plt.show()
```

```python
# Bar Charts Horizontal in variance
plt.figure(figsize = (4,3))

division = ["Div-A","Div-B","Div-C","Div-C","Div-D","Div-E"]
division_marks = [70,82,73,65,68,70]
variance = [5,8,7,6,4,2]

plt.barh(division, division_marks, xerr = variance, color = "green")
plt.title("Bar Graph", fontsize = 15, weight='bold')
plt.xlabel("division", weight='bold')
plt.ylabel("marks", weight='bold')
plt.show()
```

```python
plt.figure(figsize = (4,3))
division = ["Div-A","Div-B","Div-C","Div-C","Div-D"]
division_avarage_boys = [68,67,77,61,70]
division_avarage_girls = [72,97,69,69,50]

index = np.arange(5)
width = 0.30

plt.bar(index, division_avarage_boys, width, color = "blue", label = "Boys marks")
plt.bar(index+width, division_avarage_girls, width, color = "red", label = "Girls marks")
plt.title("Horizontally Stacked Bar Graphs", fontsize = 15, weight='bold')

plt.xlabel("divisions", weight='bold')
plt.xticks(index, division)
plt.ylabel("marks", weight='bold')

plt.legend(loc = "best", fontsize = 7)
plt.show()
```

```python
plt.figure(figsize = (4,3))
division = ["Div-A","Div-B","Div-C","Div-C","Div-D"]
division_avarage_boys = [68,67,77,61,70]
division_avarage_girls = [72,97,69,69,50]

index = np.arange(5)
width = 0.30

plt.bar(index, division_avarage_boys, width, color = "blue", label = "Boys marks")
plt.bar(index, division_avarage_girls, width, color = "red", bottom = division_avarage_boys, label = "Girls marks")

plt.title("Vertically Stacked Bar Graphs", fontsize = 15, weight='bold')
plt.xlabel("divisions", weight='bold')
plt.xticks(index, division)
plt.ylabel("marks", weight='bold')

plt.legend(loc = "best", fontsize = 7)
plt.show()
```

```python
# Pie Charts
plt.figure(figsize = (5,4))
firms = ["firm A", "firm B","firm C", "firm D", "firm E"]
market_shares = [20,25,15,10,20]
Explode = [0,0.1,0,0,0]
plt.pie(market_shares, explode=Explode, labels=firms, shadow=True, startangle=20)
plt.axis("equal")
plt.legend(fontsize=6, title="List of firms")
plt.show()
```

```python
# Histogram
x = np.random.randn(1000)
plt.figure(figsize=(4,3))
plt.title("Histogram")
plt.xlabel("Random Date")
plt.ylabel("Frequency")
plt.hist(x,10)
plt.show()
```

```python
# Scatter Plot and 3D- plotting 
sns.set_style('whitegrid')       # White Grid lines style 
plt.figure(figsize=(4,3))

height = np.array([167,170,165,155,180,166,146,159,185,145,168,172,181,169])
weight = np.array([86,74,66,78,68,79,90,73,70,88,66,84,67,84])

plt.xlim(140,200)
plt.ylim(60,100)
plt.scatter(height,weight)
plt.title("Satter Plot", weight='bold',fontsize=15)
plt.xlabel("Height", weight='bold')
plt.ylabel("Weight", weight='bold')
sns.despine()
plt.show()
```

```python
from mpl_toolkits import mplot3d
```

```python
ax = plt.axes(projection='3d')
ax.plot3D(height, weight)
ax.set_xlabel("Height", weight='bold')
ax.set_ylabel("Weight", weight='bold')
plt.show()
```
